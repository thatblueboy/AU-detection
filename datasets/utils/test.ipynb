```python
import torch
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

def visualize_graph(node_features, edge_index):
    # Convert to NumPy for easier handling
    node_features = node_features.numpy() if isinstance(node_features, torch.Tensor) else node_features
    edge_index = edge_index.numpy() if isinstance(edge_index, torch.Tensor) else edge_index
    
    # Create a graph
    G = nx.Graph()
    
    # Add nodes with positions (use first two features as x, y coordinates)
    pos = {i: (node_features[i, 0], node_features[i, 1]) for i in range(node_features.shape[0])}
    for i in range(node_features.shape[0]):
        G.add_node(i)

    # Add edges
    for src, dst in zip(edge_index[0], edge_index[1]):
        G.add_edge(int(src), int(dst))

    # Draw graph
    plt.figure(figsize=(8, 6))
    nx.draw(G, pos, with_labels=True, node_size=300, node_color='skyblue', edge_color='gray', font_size=10)
    plt.show()

# Example input (random tensors)
# node_features = torch.randn(23, 3)  # 23 nodes, 3 features per node
# edge_index = torch.randint(0, 23, (2, 74))  # Random edges (2, 74)

visualize_graph(X, index)
```
